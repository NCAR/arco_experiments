### Compute JJA anomaly for 1994-2023 w.r.t 1940-1969 for a lat-lon grid which bounds Colorado

```python
%%time
tas_monthly_zarr      = xr.open_zarr(zarr_path +'e5_tas2m_monthly_1940_2023.zarr').VAR_2T
tas_monthly_zarr
```

```python
tas_monthly_zarr      = tas_monthly_zarr.sel(latitude=slice(colorado_top,colorado_bottom),\
                                         longitude=slice(colorado_right,colorado_left))
tas_monthly_zarr 
```

```python
# %%time
# tas_monthly_nc  = xr.open_mfdataset(zarr_path +'e5_tas2m_monthly_1940_2023.nc',engine='netcdf4',chunks='auto').VAR_2T
# tas_monthly_nc  = tas_daily_nc.sel(lat=boulder_lat,lon=boulder_lon,method='nearest')
# tas_monthly_nc
```

#### Now compute monthly anomaly for JJA and comapre zarr vs netcdf

```python
months = [6,7,8]
```

```python
%%time
jja_tas_zarr = tas_monthly_zarr.sel(time=tas_monthly_zarr['time'].dt.month.isin(months))
#jja_tas_zarr
```

```python
%%time
baseline_period = np.arange(1940,1970)
current_period  = np.arange(1994,2024)
#
jja_baseline    = jja_tas_zarr.sel(time = jja_tas_zarr['time'].dt.year.isin(baseline_period))
jja_current     = jja_tas_zarr.sel(time = jja_tas_zarr['time'].dt.year.isin(current_period))
#
jja_anomaly     = jja_current.mean('time') - jja_baseline.mean('time')
jja_anomaly
```

## Plotting

```python
%%time
jja_anomaly =xr.open_zarr(zarr_path + 'e5_jja_anomaly.zarr').VAR_2T
```

```python
# Create a plot with a geographic projection
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.PlateCarree())  # Using the Plate Carree projection

# Set the extent to focus tightly on Colorado
ax.set_extent([colorado_right, colorado_left, colorado_bottom, colorado_top], crs=ccrs.PlateCarree())

# Add state borders, but only for Colorado
states = cfeature.NaturalEarthFeature(
    category='cultural',
    name='admin_1_states_provinces_lines',
    scale='50m',
    facecolor='none')

ax.add_feature(states, edgecolor='black')  # Adds only state borders

# Plot the data
jja_anomaly.plot(ax=ax, transform=ccrs.PlateCarree(),
        x='longitude', y='latitude',
        cmap='viridis',  # Choose a colormap that suits your data
        add_colorbar=True)  # Optionally add a colorbar

# Display the plot
plt.show()
```
